```python
#plotting plot
fig=plt.figure(figsize=(5,8))
X=list(range(10))
Y=[x+(x*random.random())for x in X]
plt.plot(X,Y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### Generate subplots
*subplot(number of rows,number of columns, index)*

```python
fig=plt.figure(figsize=(16,6))
ax1=plt.subplot(1,2,1)
ax1.plot(X,[x+(x*random.random()) for x in X])
ax1.set_title('Plot 1')

ax2=plt.subplot(1,2,2)
ax2.plot(X,[x-(x*random.random())for x in X])
ax2.set_title('Plot 2')
plt.show()
```

```python
fig,axes=plt.subplots(2,2)
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
axes[0][0].plot(X,[x-(x*random.random())for x in X])
axes[0][0].set_title('Plot 1')

axes[0][1].plot(X,[x+(x*random.random())for x in X])
axes[0][1].set_title('Plot 2')

axes[1][0].plot(X,[x-(x*random.random())for x in X])
axes[1][0].set_title('Plot 3')

axes[1][1].plot(X,[x+(x*random.random())for x in X])
axes[1][1].set_title('Plot 4')




```

**subplot2grid** allows to specify the object in the specific location

*subplot2grid(shape,location,rowspan,columnspan)*

```python
fig=plt.figure(figsize=(8,8))
ax1=plt.subplot2grid((4,4),(0,0),colspan=3)
ax1.plot(X,[x+(x*random.random())for x in X])
ax1.set_title('Plot1 : (0,0)')

ax1=plt.subplot2grid((4,4),(0,3))
ax1.plot(X,[x-(x*random.random())for x in X])
ax1.set_title('Plot2 : (0,3)')

ax1=plt.subplot2grid((4,4),(1,0),rowspan=3,colspan=3)
ax1.plot(X,[x-(x*random.random())for x in X])
ax1.set_title('Plot3 : (1,0)')

ax1=plt.subplot2grid((4,4),(1,3),rowspan=3,colspan=1)
ax1.plot(X,[x+(x*random.random())for x in X])
ax1.set_title('Plot4 : (1,3)')

fig.tight_layout()
plt.show()

```

### Annotate plots
##### Two ways to annotate the plots
- text()
- annotate()
<br>**Use of text()**

```python
fig=plt.figure(figsize=(6,6))
temp=[random.uniform(20,40) for i in range(5)]
city=['City A', 'City B', 'City C','City D','City E']
#my way city=["City "+i for i in list('ABCDE')]
y_pos=list(range(1,6))
graph=plt.bar(y_pos,temp,color='violet')


#xticks and more
plt.xticks(y_pos,city) #location,labels
plt.title('City Temperature')
plt.ylabel('Temperature ($^\circ$C)')

for bar,t in zip(graph,temp):
    plt.text(bar.get_x()+bar.get_width()/2.0,bar.get_height(),'%.2f $^\circ$C'%t,ha='center',va='bottom')
    #ha-horizontal alignment, va-vertical alignment

plt.show()
```

### Use of annotate() function

```python
fig=plt.figure(figsize=(8,6))
plt.plot(X,np.exp(X))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Annotating Exponential Plot using plt.annotate()')
plt.annotate('Point 1',xy=(6,400),arrowprops=dict(arrowstyle='->'),xytext=(4,600))
#xytext=positon of text defatult xy
plt.annotate('Point 2',xy=(7,1150),arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-.2'),xytext=(4.5,2000))
plt.annotate('Point 2',xy=(8,3000),arrowprops=dict(arrowstyle='-|>',connectionstyle='angle,angleA=90,angleB=0'),xytext=(8.5,2200))




plt.show()
```

## Group bar plots

```python
temp_summer=[random.uniform(20,40) for i in range(5)]
temp_winter=[random.uniform(0,10) for i in range(5)]
fig=plt.figure(figsize=(10,5))
city=['City A','City B','City C','City D','City E']
x_pos_summer=list(range(1,6))
x_pos_winter=[i+0.4 for i in x_pos_summer]

#plot
graph_summer=plt.bar(x_pos_summer,temp_summer,color='tomato',label='Summer',width=0.4)
graph_winter=plt.bar(x_pos_winter,temp_winter,color='dodgerblue',label='Winter',width=0.4)

#ticks
plt.xticks([i+0.2 for i in x_pos_summer],city)
plt.title('City Temperature')
plt.ylabel('Temperatue ($^\circ$C)')

#Annotating graphs
for summer_bar,winter_bar,ts,tw in zip(graph_summer,graph_winter,temp_summer,temp_winter):
    plt.text(summer_bar.get_x()+summer_bar.get_width()/2.0,summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x()+winter_bar.get_width()/2.0,winter_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
   
    
plt.legend()
plt.show()
    
```

# Modify Ticks

```python
fig=plt.figure(figsize=(10,6))
temp_summer=[random.uniform(20,40) for i in range(5)]
temp_winter=[random.uniform(0,10) for i in range(5)]
fig=plt.figure(figsize=(10,5))
city=['City A','City B','City C','City D','City E']
x_pos_summer=list(range(1,6))
x_pos_winter=[i+0.4 for i in x_pos_summer]

#plot
graph_summer=plt.bar(x_pos_summer,temp_summer,color='tomato',label='Summer',width=0.4)
graph_winter=plt.bar(x_pos_winter,temp_winter,color='dodgerblue',label='Winter',width=0.4)

#ticks
plt.xticks([i+0.2 for i in x_pos_summer],city)
plt.title('City Temperature')
#plt.ylabel('Temperatue ($^\circ$C)')
#removing axes from the figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

#Modifying the ticks
plt.xticks([i+0.2 for i in x_pos_summer],city,fontname='Chilanka',rotation=45,fontsize=14)


#Annotating graphs
for summer_bar,winter_bar,ts,tw in zip(graph_summer,graph_winter,temp_summer,temp_winter):
    plt.text(summer_bar.get_x()+summer_bar.get_width()/2.0,summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x()+winter_bar.get_width()/2.0,winter_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
   
    
plt.legend()
plt.show()
    
```

# Modify Legend
It is done using the legend() function with various arguments such as **loc** **frameon** **ncol**

```python
fig=plt.figure(figsize=(10,6))
temp_summer=[random.uniform(20,40) for i in range(5)]
temp_winter=[random.uniform(0,10) for i in range(5)]
fig=plt.figure(figsize=(10,5))
city=['City A','City B','City C','City D','City E']
x_pos_summer=list(range(1,6))
x_pos_winter=[i+0.4 for i in x_pos_summer]

#plot
graph_summer=plt.bar(x_pos_summer,temp_summer,color='tomato',label='Summer',width=0.4)
graph_winter=plt.bar(x_pos_winter,temp_winter,color='dodgerblue',label='Winter',width=0.4)

#ticks
plt.xticks([i+0.2 for i in x_pos_summer],city)
plt.title('City Temperature')
plt.ylabel('Temperatue ($^\circ$C)')

#Annotating graphs
for summer_bar,winter_bar,ts,tw in zip(graph_summer,graph_winter,temp_summer,temp_winter):
    plt.text(summer_bar.get_x()+summer_bar.get_width()/2.0,summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x()+winter_bar.get_width()/2.0,winter_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
   
#modify legned
    
plt.legend(loc='upper center',ncol=2,frameon=False)
plt.show()
    
```

# Add Watermark

```python
fig=plt.figure(figsize=(10,6))
temp_summer=[random.uniform(20,40) for i in range(5)]
temp_winter=[random.uniform(0,10) for i in range(5)]
fig=plt.figure(figsize=(10,5))
city=['City A','City B','City C','City D','City E']
x_pos_summer=list(range(1,6))
x_pos_winter=[i+0.4 for i in x_pos_summer]

#plot
graph_summer=plt.bar(x_pos_summer,temp_summer,color='tomato',label='Summer',width=0.4)
graph_winter=plt.bar(x_pos_winter,temp_winter,color='dodgerblue',label='Winter',width=0.4)

#ticks
plt.xticks([i+0.2 for i in x_pos_summer],city)
plt.title('City Temperature')
plt.ylabel('Temperatue ($^\circ$C)')

#Text WaterMark
fig.text(0.85,0.15,"Analytical Vidhya",fontsize=65,color='gray',ha='right',va='bottom',alpha=0.4,rotation=14)

#Annotating graphs
for summer_bar,winter_bar,ts,tw in zip(graph_summer,graph_winter,temp_summer,temp_winter):
    plt.text(summer_bar.get_x()+summer_bar.get_width()/2.0,summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x()+winter_bar.get_width()/2.0,winter_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
   
#modify legned
    
plt.legend(loc='upper center',ncol=2,frameon=False)
plt.show()
    
```
